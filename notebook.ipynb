## Quality-Differentiation

```python
# ----------------------------------------
# Quality Differentiation Heatmap Visualization
# ----------------------------------------

# Define parameter ranges for the axes
THETA_RANGE = (1, 5)         # X-axis: Marginal Search Cost (θ)
Q_H_RATIO_RANGE = (2, 5)     # Y-axis: Quality Ratio (q_H / q_L)

# Fixed market parameters
FIXED_PARAMS = {
    'Delta': 0.75,   # Buy Box Bias (Δ)
    'q_L': 10,       # Base demand for low-quality products
    'c': 5,          # Marginal cost
    'N': 3,          # Total number of sellers
    'K': 1           # Number of high-quality sellers
}

# Number of sellers in the buy box to consider
M_VALS = [1, 2, 3]

# Resolution of the heatmap grid
RESOLUTION = 100

# Generate the heatmap data for quality differentiation
heatmap_df = _generate_heatmap_data_q_ratio(
    theta_range=THETA_RANGE,
    q_h_ratio_range=Q_H_RATIO_RANGE,
    m_vals=M_VALS,
    fixed_params=FIXED_PARAMS,
    resolution=RESOLUTION
)

# Prepare the q_H/q_L values for plotting
q_h_ratio_vals_for_plot = np.linspace(Q_H_RATIO_RANGE[0], Q_H_RATIO_RANGE[1], RESOLUTION)

# Plot the welfare and profit heatmaps as a function of quality differentiation
plot_welfare_heatmap_q_ratio(
    df=heatmap_df,
    m_vals=M_VALS,
    q_h_ratio_vals=q_h_ratio_vals_for_plot,
    fixed_params=FIXED_PARAMS
)
```
